```python
#there are 5 subjects psychology,nanotechnology,medicine,oncology,and anthropology.two students have applied to do master in any of these 5 fields with a line plot show in which feild they can excel.the cutoff for every filed is 70
#student1=[75,45,73,71,57]
#student2=[43,67,68,53,85]
student1=[75,45,73,71,57]
subjects=['psychology','nano technology','medicine','oncology','anthropology']
student2=[43,67,68,53,85]
cutoff=[70,70,70,70,70]
#student1
#plt.plottype(X,Y,label,color,marker,linewidth,linestyle)
plt.plot(subjects,student1,label="student1")
#student2
plt.plot(subjects,student2,label="student2")
#cutoff
plt.plot(subjects,cutoff,label="cutoff")
plt.title("marks")
plt.xlabel("subjects")
plt.ylabel("marks")
plt.legend()
```

```python
#there are 5 subjects psychology,nanotechnology,medicine,oncology,and anthropology.two students have applied to do master in any of these 5 fields with a line plot show in which feild they can excel.the cutoff for every filed is 70
#student1=[75,45,73,71,57]
#student2=[43,67,68,53,85]
student1=[75,45,73,71,57]
subjects=['psychology','nano technology','medicine','oncology','anthropology']
student2=[43,67,68,53,85]
cutoff=[70,70,70,70,70]
#student1
plt.scatter(subjects,student1,label="student1",color='k',marker="p")         #labelling
#student2
plt.scatter(subjects,student2,label="student2",color='b',marker="H")           #coloring
#cutoff
plt.plot(subjects,cutoff,label="cutoff",color='r',linestyle=":",linewidth=4)
plt.title("marks")
plt.xlabel("subjects")
plt.ylabel("marks")
plt.legend()
```

```python
#An india vs australila match is on live india has given 298 as target for australila. the runs for every 5 overs is given as follows
#india=[25,51,84,131,160,189,220,250,267,297]
#australila have completed 25 overs
#aus=[15,41,94,110,151]
#ploat overs Vs runs for both the contries with india as a blue line and australila as yellow line.give appropriate labels and legend.the marker for india is 'flag' and marker for austrila is 'plus',take overs from arange funciton ,creates runs as array.
import numpy as np
india=np.array([25,51,84,131,160,189,220,250,267,297])
australila=np.array([15,41,94,110,151])
oversa=np.arange(5,26,5)
oversi=np.arange(5,51,5)
plt.plot(oversi,india,label="india",color='b',marker=">",linewidth=3)
plt.plot(oversa,australila,label="aus",color='y',marker="+",linewidth=3                                                                )
plt.title("match")
plt.xlabel("runs")
plt.ylabel("overs")
plt.legend()
```
