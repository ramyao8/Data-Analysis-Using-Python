```python
#create a list of 10 integer elements print their squares save these squares in new list
l=[1,10,12,345,23,67,123,78,988,86]
l2=[]
for i in l:
  l2.append(i**2)
print(l2)
```

```python
#list comprehension
#newlist=[output for if ]
l1=[1,10,12,345,23,67,123,78,988,86]
l2=[i*i for i in l1 if i>50]
print(l2)
```

```python
#list comprehension
#newlist=[(st of if) if (cond) else (st of else) for  ]
l1=[1,10,12,345,23,67,123,78,988,86]
l2=[i*i if i>50 else 'x' for i in l1]
print(l2)
```

```python
#numpy
import numpy as np
A=np.array([1,2,3,4])   #1-d array
print(A.ndim)           # (columns,)
print(A.shape)
```

```python
B=np.array([[1,2,3],[4,5,6]]) #2-d array
#elements in each row=columns
print(B)
print(B.ndim)   #2-d array
print(B.shape)  # (rows,columns)
```

```python
#creating 3-d array
C=np.array([[[1,2,3],[4,5,6]],[[11,12,13],[56,76,6]]])
print(C)
print(C.ndim)   #3-d array
print(C.shape)  #(groups,rows,columns)   #no.of elements=groups*rows*columns
```
